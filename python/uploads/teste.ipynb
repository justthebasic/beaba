```python
import pandas as pd

data = {
    "nome_campo": [
        "Nome",
        "Idade",
        
    ],
    "tipo_campo": [
        "object",
        "int64",
        
    ]
}
type_mapping = {
    'object': object,  # Correção do mapeamento para corresponder aos tipos do DataFrame
    'int64': int,
    'float64': float,
    'bool': bool,
    # Adicione mais mapeamentos conforme necessário
}



df = pd.read_excel('Maria (cópia).xlsx' )

display(df)

# Função para validar o DataFrame em relação ao template
def validar_dataframe(df, data, type_mapping):
    campo_names = data["nome_campo"]
    campo_tipos = data["tipo_campo"]

    for col in df.columns:
        col_name = col
        col_dtype = df[col_name].dtype
        expected_dtype = type_mapping[campo_tipos[campo_names.index(col_name)]]
        if col_dtype != expected_dtype:
            raise ValueError(f"O tipo de dados da coluna '{col_name}' não corresponde ao tipo esperado.")

try:
    validar_dataframe(df, data, type_mapping)
    print("O DataFrame foi validado com sucesso!")
except ValueError as e:
    print(f"Erro de validação: {str(e)}")
```

```python
import pandas as pd
from faker import Faker

# Criar uma instância do Faker
fake = Faker()

# Especificar o caminho do arquivo CSV existente
caminho_arquivo_existente = 'Pedido.csv'

# Ler o arquivo CSV existente
df_existente = pd.read_csv(caminho_arquivo_existente)

# Exibir o DataFrame antes da adição de novos dados
print(f'Número de linhas antes da expansão: {len(df_existente)}')

# Número desejado de linhas
num_linhas_desejado = 10


# Determinar quantas vezes repetir o DataFrame existente para atingir o número desejado
num_repeticoes = num_linhas_desejado // len(df_existente) + 1

# Repetir o DataFrame existente
df_repetido = pd.concat([df_existente] * num_repeticoes, ignore_index=True)

# Preencher colunas com dados fictícios usando Faker
df_repetido['ID do Pedido'] = [fake.id() for _ in range(len(df_repetido))]
df_repetido['Produto'] = [fake.word() for _ in range(len(df_repetido))]
df_repetido['Quantidade'] = [fake.random_int(min=1, max=100) for _ in range(len(df_repetido))]
df_repetido['Endereço de Entrega'] = [fake.address() for _ in range(len(df_repetido))]
df_repetido['Data de Entrega Prevista'] = [fake.date_between(start_date='+1d', end_date='+30d').strftime('%Y-%m-%d') for _ in range(len(df_repetido))]
df_repetido['Status do Pedido'] = [fake.random_element(elements=('Em Processamento', 'Enviado', 'Entregue')) for _ in range(len(df_repetido))]

# Adicione mais colunas conforme necessário

# Amostrar aleatoriamente para obter exatamente o número desejado de linhas
df_atualizado = df_repetido.sample(n=num_linhas_desejado, replace=True)

# Exibir o DataFrame após a adição de novos dados
print('\nDataFrame atualizado:')
print(df_atualizado)

# Salvar o DataFrame atualizado de volta no mesmo arquivo CSV
df_atualizado.to_csv(caminho_arquivo_existente, index=False)

print(f'\nDados adicionados e arquivo atualizado em {caminho_arquivo_existente}')
```
