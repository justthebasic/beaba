```python
import pandas as pd

data = {
    "nome_campo": [
        "Nome",
        "Idade",
        
    ],
    "tipo_campo": [
        "object",
        "int64",
        
    ]
}
type_mapping = {
    'object': object,  # Correção do mapeamento para corresponder aos tipos do DataFrame
    'int64': int,
    'float64': float,
    'bool': bool,
    # Adicione mais mapeamentos conforme necessário
}



df = pd.read_excel('Maria.xlsx',header=0,skiprows=1 )

display(df)

# Função para validar o DataFrame em relação ao template
def validar_dataframe(df, data, type_mapping):
    campo_names = data["nome_campo"]
    campo_tipos = data["tipo_campo"]

    for col in df.columns:
        col_name = col
        col_dtype = df[col_name].dtype
        expected_dtype = type_mapping[campo_tipos[campo_names.index(col_name)]]
        if col_dtype != expected_dtype:
            raise ValueError(f"O tipo de dados da coluna '{col_name}' não corresponde ao tipo esperado.")

try:
    validar_dataframe(df, data, type_mapping)
    print("O DataFrame foi validado com sucesso!")
except ValueError as e:
    print(f"Erro de validação: {str(e)}")
```
